```python
from scipy.signal import butter, filtfilt


def bandpass_filter(data, lowcut=30, highcut=100, fs=256, order=5):
    """
    Apply a band-pass filter to the input signal.
    
    Parameters:
        data (array): Input signal.
        lowcut (float): Low cutoff frequency (Hz).
        highcut (float): High cutoff frequency (Hz).
        fs (float): Sampling frequency (Hz).
        order (int): Order of the filter.
    
    Returns:
        array: Filtered signal.
    """
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    
    b, a = butter(order, [low, high], btype='band')
    filtered_data = filtfilt(b, a, data)
    
    return filtered_data
```

```python
# Step 3: Project data to Stiefel manifold using QR decomposition
def project_to_stiefel(data):
    stiefel_data = []
    for trial in data:
        Q, R = np.linalg.qr(trial)  # QR decomposition works on (795 x 64)
        stiefel_data.append(Q)
    return np.array(stiefel_data)

stiefel_matrices = project_to_stiefel(transposed_data)
```

```python
import numpy as np
from scipy.linalg import svd

def riemannian_mean_stiefel(data, max_iter=1000, tol=1e-6):
    """
    Compute the Fréchet mean on the Stiefel manifold and check for convergence.
    """
    mean = data[0].reshape(-1, 1)  # Ensure mean starts as a 2D matrix
    for iteration in range(max_iter):
        tangent_vectors = np.array([log_map_stiefel(mean, data[i].reshape(-1, 1)) for i in range(data.shape[0])])
        tangent_mean = np.mean(tangent_vectors, axis=0)
        mean_new = exp_map_stiefel(mean, tangent_mean)

        # Compute the difference norm for convergence check
        diff = np.linalg.norm(mean_new - mean)
        print(f"Iteration {iteration + 1}: Convergence difference = {diff}")

        if diff < tol:
            print(f"Converged at iteration {iteration + 1} with difference {diff}.")
            return mean_new  # Converged!
        
        mean = mean_new
    
    print("Warning: Algorithm did not converge within the maximum number of iterations.")
    return mean

def log_map_stiefel(base, point):
    """
    Compute the Riemannian Logarithm map on the Stiefel manifold.
    """
    base = base.reshape(-1, 1)  # Ensure 2D
    point = point.reshape(-1, 1)  # Ensure 2D
    
    M = np.dot(base.T, point)
    skew = 0.5 * (M - M.T)
    tangent_vector = point - np.dot(base, M) + np.dot(base, skew)
    return tangent_vector

def exp_map_stiefel(base, tangent):
    """
    Compute the Riemannian Exponential map on the Stiefel manifold.
    """
    base = base.reshape(-1, 1)  # Ensure 2D
    tangent = tangent.reshape(-1, 1)  # Ensure 2D
    if base.ndim != 2 or tangent.ndim != 2:
        raise ValueError(f"Expected 2D matrices, got base {base.shape} and tangent {tangent.shape}")
    
    U, _, Vt = svd(base + tangent, full_matrices=False)
    return np.dot(U, Vt)

def process_stiefel_data(train_data, test_data):
    """
    Compute the Fréchet mean for each channel and map all trials to the tangent space.
    """
    num_trials_train, num_samples, num_channels = train_data.shape
    num_trials_test = test_data.shape[0]
    
    train_tangent_space = np.zeros_like(train_data)
    test_tangent_space = np.zeros_like(test_data)
    
    for ch in range(num_channels):
        print(f"Processing Channel {ch+1}/{num_channels}...")
        
        train_channel_data = train_data[:, :, ch]
        test_channel_data = test_data[:, :, ch]
        
        if train_channel_data.ndim != 2:
            raise ValueError(f"train_channel_data should be 2D, got shape {train_channel_data.shape}")
        
        frechet_mean = riemannian_mean_stiefel(train_channel_data)
        
        for i in range(num_trials_train):
            train_tangent_space[i, :, ch] = log_map_stiefel(frechet_mean, train_channel_data[i, :].reshape(-1, 1)).flatten()
        
        for i in range(num_trials_test):
            test_tangent_space[i, :, ch] = log_map_stiefel(frechet_mean, test_channel_data[i, :].reshape(-1, 1)).flatten()
    
    return train_tangent_space, test_tangent_space

# Example usage:
# train_tangent, test_tangent = process_stiefel_data(train_data, test_data)
```
